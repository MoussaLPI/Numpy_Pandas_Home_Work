# Packages required

```python
import pandas as pd
import os
from collections import Counter
```

```python
files=os.listdir('./the_boys/') # List all files from current directory passed in parameter
```

```python
# Number of season

number_season=[ int(file[2]) for file in files]

Counter(number_season) #Count number of episodes in seasons
```

## We have three seasons with 9 episode for season 1, 8 and 8 for season 2 and 3

```python
#Cancatening DataFrames

df=pd.DataFrame()

for file in files:
    
    boy=pd.read_csv('./the_boys/'+file)
    
    boy['Season']=int(file[2])
    boy['Episode']=int(file[-5])
    
    df=pd.concat([df,boy])
```

# Most appeared character in serie

```python
df['time_apparition']=df['end']-df['start'] # Duration of characters by passage
```

```python
df_grouped_by_nconst=df.groupby(['nconst','character']).sum().sort_values(['time_apparition'],ascending=False)
df_grouped_by_nconst=df_grouped_by_nconst.iloc[:10,:]

df_grouped_by_nconst=pd.DataFrame(df_grouped_by_nconst)
df_grouped_by_nconst['time_apparition'].plot(kind='bar',x='character')
```

# Most 10 famous character in season 1

```python
df_so1=df[df['Season']==1]


df_so1=df_so1.groupby(['nconst','character']).sum().sort_values(['time_apparition'],ascending=False)
df_so1=df_so1.iloc[:10,:]

df_so1=pd.DataFrame(df_so1)
df_so1['time_apparition'].plot(kind='bar',x='character')
```
