```python
#import necessery libraries
import pandas as pd
import numpy as np
import requests
```

```python
#store our data in DataFrame Object and normalize it
data=requests.get('http://app-sellaci.eu-central-1.elasticbeanstalk.com/')
pd1=data.json()
df=pd.json_normalize(pd1)

#show all columns
pd.set_option("display.max_columns", None)
print(df.shape)
df
```

```python
#get info for the data
df.info()
```

```python
#describe data
df.describe()
```

```python
#find the sum of all null values in columns
df.isnull().sum()
```

```python
#replace all false values and unknown values by nan to easy drop this column that full in nan values
df=df.replace([False,'unknown',''],np.nan)
df
```

```python
# The id for the second row is obviously a typo
df["_id"]=df["_id"].str.replace("23232323","")
df
```

```python
#drop all columns which contain nan values in all ros
df = df.dropna(how='all',axis=1)
df
```

```python
df.shape
```

```python
# Here we drop the columns where the data has identical values. These attributes do not provide any comparison
keys= []
values = []
for column_i in df:
    if type(df[column_i].iloc[0]) != list:
        set_i = set(df[column_i])
        if len(set_i) == 1:
            keys.append(column_i)
            values.append(list(set_i)[0])


df=df.drop(columns=keys)

identical_columns = dict(zip(keys,values))
print(identical_columns)
print("These columns are the same for all rows. We don't show them in dataframes")
```
